```python

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel(r"D:\PythonDescri-Project level\NewDS.xlsx")
df.head()
```

```python
sort=df.sort_values(by= 'Price', ascending=True)
c=sns.color_palette("Set1")
bars=plt.barh(df['Model'], sort['Price'], height=0.5, color=c, alpha=0.9)

sns.set(rc={'figure.figsize': (15,20)})
plt.grid(False)

plt.xlabel("Price(INR)", fontsize=20)
plt.ylabel("Model Name" ,fontsize=20)
plt.title("Porsche Price Analysis", fontsize=21)

df_sorted = df.sort_values(by='Price', ascending=True)
for i, bar in enumerate(bars):
   plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{df_sorted["Price"].iloc[i]:.2f} Cr', va='center', ha='left', fontsize=15)


plt.show()
```

```python
plt.title("Revenue - Porsche", fontsize = 25)
filtered_data = df.dropna(subset=['Revenue', 'Year'])

plt.axis('equal') 
colors= ['crimson','g','b','orange']
labels=filtered_data['Year'].astype(int)
explode = (0.0, 0.0, 0., 0.0)

# Plot pie chart with sales data and year labels
width=2
height=2
plt.pie(filtered_data['Revenue'], labels = labels, colors = colors, explode = explode, autopct='%0.1f%%', startangle= 90, counterclock=False,
         radius=1, shadow=True)
plt.pie([1],colors = "white", radius=0.4)

sns.set(rc={'figure.figsize': (4,4)})
plt.legend(loc='upper left', bbox_to_anchor=(1.5,1))
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()

    
```

```python
labels=filtered_data['Year'].astype(int)
labels2=filtered_data['Sale'].astype(int)

plt.plot(labels,labels2,color='r',marker='o')

plt.xlabel("Year",fontsize=21,color='Green')
plt.ylabel("Sale", fontsize=21, color='Green')

plt.xticks(color='blue')
plt.yticks(color='blue')

plt.title("Porsche - Cars Delivered", fontsize=21, color='Black')

plt.grid(False)
plt.show()
```

```python
plt.title("Operating Profit", fontsize = 25)
filtered_data = df.dropna(subset=['Profit', 'Year'])
colors= sns.color_palette("Set3")
labels=filtered_data['Year'].astype(int)
explode = (0.0, 0.0, 0.0, 0.1)

# Plot pie chart with sales data and year labels
plt.pie(filtered_data['Profit'], labels = labels, colors = colors, explode = explode, autopct='%1.1f%%', startangle= 90, counterclock=False, shadow = True)
plt.legend(loc='upper left', bbox_to_anchor=(1.5,1))
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
mile = df.groupby('Model Series')['mileage'].sum().reset_index()
# Creating the bar plot
sns.barplot(data=mile, x='Model Series', y='mileage', palette="Set3")

plt.title("Porsche Mileage")
plt.xlabel("Model Series")
plt.ylabel("Mileage")

plt.xticks(rotation=90)  # Rotating x-axis labels for better readability
plt.figure(figsize = (5,2))


plt.show()
```

```python
data = df['MS']
sns.countplot(x=data, width=0.4, palette="Set1")   #each data
plt.figure(figsize=(4, 4))


plt.show()
```
